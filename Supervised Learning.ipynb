```python
#Part 2.3 a- Split data into X and Y

# Drop 'ID' and 'ZipCode' columns from the DataFrame
merged_bankdt.drop(['ID', 'ZipCode'], axis=1, inplace=True)

# Split the data into feature variables (X) and the target variable (Y)
X = merged_bankdt.drop('LoanOnCard', axis=1)  # X contains all columns except 'LoanOnCard'
Y = merged_bankdt['LoanOnCard']              # Y contains only the 'LoanOnCard' column

# Display the shapes of X and Y to verify the split
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
#Part 2.3 b - Split data into train and test. Keep 25% data reserved for testing

from sklearn.model_selection import train_test_split

# Split the data into training (75%) and testing (25%) sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

# Print the shapes of the training and testing sets to verify the split
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

```python
#Part 2.3 E - Balance the data using the right balancing technique.

# i. Check distribution of the target variable
print("\nOriginal Distribution of Target Variable:")
print(merged_bankdt['LoanOnCard'].value_counts())

from imblearn.over_sampling import SMOTE

# ii. Get the unique class labels in the 'LoanOnCard' column
unique_classes = merged_bankdt['LoanOnCard'].unique()

# iii. Here you need to balance the target variable as 50:50

# Remove 'nan' class from unique_classes (optional, if needed)
unique_classes = [cls for cls in unique_classes if pd.notnull(cls)]

# Calculate the percentage of each class in the original data
class_a_percentage = (merged_bankdt['LoanOnCard'].value_counts()[unique_classes[0]] / len(merged_bankdt)) * 100
class_b_percentage = (merged_bankdt['LoanOnCard'].value_counts()[unique_classes[1]] / len(merged_bankdt)) * 100
print("\nClass A Percentage:", class_a_percentage)
print("Class B Percentage:", class_b_percentage)

# Create a dictionary specifying the desired class distribution based on the unique classes
class_distribution = {unique_classes[0]: len(merged_bankdt) // 2, unique_classes[1]: len(merged_bankdt) // 2}

# Instantiate SMOTE with the dictionary
smote = SMOTE(sampling_strategy=class_distribution, random_state=42)

# Separate X (features) and Y (target variable) from the DataFrame
X = merged_bankdt.drop(['LoanOnCard'], axis=1)
Y = merged_bankdt['LoanOnCard']

# Apply SMOTE to the target variable (Y) only (to avoid data leakage)
X_resampled, Y_resampled = smote.fit_resample(X, Y)

# Check the new distribution of the target variable after applying SMOTE
print("\nResampled Distribution of Target Variable:")
print(pd.Series(Y_resampled).value_counts())


#iv. Try appropriate method to achieve the same
#Used Smote for the same
```

```python
#Part 2.3 F - Again train the same previous model on balanced data

# Separate X (features) and Y (target variable) from the DataFrame
X = merged_bankdt.drop(['LoanOnCard'], axis=1)
Y = merged_bankdt['LoanOnCard']

# Split the data into training (75%) and testing (25%) sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

# Create a Logistic Regression model with class_weight parameter to handle imbalance
logreg_model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)

# Train the model on the training data
logreg_model.fit(X_train, Y_train)
```

```python
#Part 2.3 G - Print evaluation metrics and clearly share differences observed

from sklearn.metrics import confusion_matrix

# Step 1: Evaluate the original model on the imbalanced test set
Y_pred_imbalanced = logreg_model.predict(X_test)

# Calculate evaluation metrics for the imbalanced model
accuracy_imbalanced = accuracy_score(Y_test, Y_pred_imbalanced)
report_imbalanced = classification_report(Y_test, Y_pred_imbalanced)
conf_matrix_imbalanced = confusion_matrix(Y_test, Y_pred_imbalanced)

# Print evaluation metrics for the imbalanced model
print("Imbalanced Model Evaluation:")
print("Accuracy:", accuracy_imbalanced)
print("Classification Report:\n", report_imbalanced)
print("Confusion Matrix:\n", conf_matrix_imbalanced)

# Step 2: Evaluate the new model on the balanced test set
Y_pred_balanced = logreg_model.predict(X_resampled)

# Calculate evaluation metrics for the balanced model
accuracy_balanced = accuracy_score(Y_resampled, Y_pred_balanced)
report_balanced = classification_report(Y_resampled, Y_pred_balanced)
conf_matrix_balanced = confusion_matrix(Y_resampled, Y_pred_balanced)

# Print evaluation metrics for the balanced model
print("\nBalanced Model Evaluation:")
print("Accuracy:", accuracy_balanced)
print("Classification Report:\n", report_balanced)
print("Confusion Matrix:\n", conf_matrix_balanced)

```
